```python
# Monting File & Read. Also importing library - Pandas and Numpy

from google.colab import drive
import pandas as pd
import numpy as np

drive.mount('/content/drive')

file_path = ('/content/drive/My Drive/Data Analyst/Projects/Raw Data/Online Sales Data.csv')

df = pd.read_csv(file_path)
```

```python
#Checking first few rows
print(df.head())
```

```python
#Checking last few rows
print(df.tail())
```

```python
#Checking data types and summary of the dataframe
print(df.info())
```

```python
#Checking descriptive statistics summary of the dataframe
print(df.describe())
```

```python
#Checking for missing values
print(df.isnull().sum())
```

```python
#cheking conditional rows as per index
rows = df.iloc[1:5]
print(rows)
```

```python
#Dataframe slicing
rows2 = df[10:14]
print(rows2)
```

```python
# Column slicing
rows2 = df['Product Name'][10:14]
print(rows2)
```

```python
# 2 columns slicing & indexing by position
columns = df[['Product Category', 'Product Name']].iloc[2:5]
print(columns)
```

```python
# slicing & indexing by label
columns = df.loc[1:6, ['Region']]
print(columns)
```

```python
#Renaming column names
df=df.rename(columns={'Date': 'Dates'})
print(df.head())
```

```python
#Droping duplicates if any

df = df.drop_duplicates()
```

```python
#Product Groupby revenue
grouped = df.groupby('Product Category')
category_revenue= grouped['Total Revenue'].agg('sum')
#category_revenue = category_revenue.rename('Category Profit')
category_revenue = category_revenue.sort_values(ascending=False)
print(category_revenue)
```

```python
print(df.info())
```

```python
#Converting date column datatypes to datetime & extrasct year, month & date
df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
print(df.head())
```

```python
#filtering total revenue & check dataframe for selected columns

mx_rev = df[df['Total Revenue'] > 1500]
print(mx_rev[['Product Category', 'Total Revenue', 'Product Name', 'Region', 'year', 'month']])
```

```python
# Pivot Table created Product Category wise
pivot = df.pivot_table(index='Product Category', columns='Region',values='Total Revenue', aggfunc='sum')
print(pivot)
```

```python
#Sorting data
sort_df = df.sort_values(by='Total Revenue')
print(sort_df[['Total Revenue', 'Product Category' ]][0:10])
```

```python
#Sorting data & resetting index
sort_df = df.sort_values(by='Total Revenue')
sort_df = sort_df.reset_index(drop=True)
print(sort_df[['Total Revenue', 'Product Category' ]][0:10])
```

```python
#Transposing the dataframe and seein 2 columns only
transposed_df = df.T
print(transposed_df[[0,1]])
```
